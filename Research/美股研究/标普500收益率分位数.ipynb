```python
from WindPy import w
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
data = w.wsd('SPX.GI', 'close', '2000-01-01', '2019-12-24')
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['SPX.GI'])
df.index = pd.to_datetime(df.index)
```

```python
window = 121
ret = df['SPX.GI'].pct_change(window)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot(ret.resample('W').last().dropna(), color='blue', label=u'标普500 %d个月收益率'%(window // 20))
axx = ax.twinx()
axx.plot(ret.rank(pct=True).resample('W').last().dropna(), color='red', label=u'标普500 %d个月收益率分位数（右轴）'%(window // 20))

ax.legend(fontsize=15, loc='upper left')
axx.legend(fontsize=15, loc='upper right')
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
axx.yaxis.set_tick_params(labelsize=20)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals])
vals = axx.get_yticks()
axx.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals])

_ = ax.set_title(u'标普500 %d个月收益率'%(window//20), fontsize=25, y=1.02)
```
