```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ind_df = pd.read_excel('D:/data/index/881001.WI.xlsx')
```

```python
cap_df = pd.read_excel('D:/market cap.xlsx')
```

```python
cap_df = cap_df.transpose()
```

```python
p_df = pd.read_excel('D:/market.xlsx')
```

```python
p_df = p_df.transpose()
```

```python
p_df.tail()
```

# 000631.SZ是个异常数据

```python
cols = p_df.columns.difference(['000631.SZ'])
```

```python
r_df = r_df[cols]
p_df = p_df[cols]
```

# 计算Dispersion

```python
r_df = p_df.pct_change()
```

```python
exr_df = np.exp(r_df)
```

```python
m_df = cap_df.div(cap_df.sum(axis=1), axis='index')
```

```python
m_df = m_df.replace([np.inf, -np.inf], np.NAN)
r_df = r_df.replace([np.inf, -np.inf], np.NAN)
exr_df = exr_df.replace([np.inf, -np.inf], np.NAN)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(ind_df['close'])
ax2 = ax1.twinx()
t = np.log((m_df * exr_df).sum(axis=1))
ax2.bar(t.index, t, color='red', width=15)
```

```python
y = ind_df['close'].resample('M').last().pct_change(1).shift(-1)
plt.figure(figsize=(16, 9))
plt.scatter(t, y)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(ind_df['close'])
ax2 = ax1.twinx()
ax2.plot((m_df * r_df).sum(axis=1), color='red')
```

```python
d = np.log((m_df * exr_df).sum(axis=1)) - (m_df * r_df).sum(axis=1)
```

```python
d = d.replace([np.inf, -np.inf], np.NAN)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(ind_df['close'])
ax2 = ax1.twinx()
# ax2.plot(d, color='red')
ax2.bar(d.index, d, color='red', width=15)
```

```python
rank_d = d.rolling(window=120).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(ind_df['close'])
ax2 = ax1.twinx()
ax2.plot(rank_d, color='red')
```
