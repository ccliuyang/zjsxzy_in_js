```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DATA_DIR = 'D:/Data/stocks/'
```

```python
ind = pd.read_excel('D:/Data/index/000905.SH.xlsx')
```

```python
comp = pd.read_excel('D:/Data/index-component/000905.SH.xlsx')
```

```python
comp.head()
```

```python
dic = {}
for code in comp['wind_code']:
    fname = '%s/%s.xlsx'%(DATA_DIR, code)
    temp = pd.read_excel(fname)
    dic[code] = temp[['close', 'volume']]
```

```python
pnl = pd.Panel(dic)
```

```python
pnl.loc[:, :, 'ret'] = pnl.minor_xs('close').pct_change(10)
```

```python
mask = pnl.minor_xs('ret').copy()
```

```python
mask[mask > 0] = 1
mask[mask < 0] = 0
```

```python
up = mask.sum(axis=1)
up_amt = (mask * pnl.minor_xs('volume').rolling(10).sum()).sum(axis=1)
```

```python
mask = pnl.minor_xs('ret').copy()
mask[mask > 0] = 0
mask[mask < 0] = 1
```

```python
down = mask.sum(axis=1)
down_amt = (mask * pnl.minor_xs('volume').rolling(10).sum()).sum(axis=1)
```

```python
ratio = ((up+1) / (down+1)) / (up_amt / down_amt)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(ind[ind.index >= '2010-01-01']['close'], color='red')
ax2 = ax1.twinx()
# ax2.plot(down.rolling(100).mean(), color='red')
ax2.plot(ratio, color='blue')
```
