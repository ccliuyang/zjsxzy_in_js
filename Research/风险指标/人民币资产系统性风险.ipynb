```python
import pandas as pd
import numpy as np
import math
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
from WindPy import w
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel('D:/Data/risk/systematic_data.xlsx')
```

```python
const = pd.read_excel('D:/Data/risk/systematic_const.xlsx')
```

```python
const
```

# 人民币资产系统性风险

```python
ch = df[const['China'].dropna()]
```

```python
ch_risk = pd.Series(index=ch.index)
window = 243
for date in ch.index[window:]:
    past = ch[ch.index <= date][-window:]
    if past.shape[0] == window:
        past = past.dropna(axis=1)
        if past.shape[1] >= 5:
            k = int(math.ceil(past.shape[1] / 5))
            x = StandardScaler().fit_transform(past)
            pca = PCA()
            pca.fit(x)
            pricipal = pca.explained_variance_ratio_
#             print ch_risk.loc[date], k
            ch_risk.loc[date] = pricipal[:k].sum()
```

```python
ch_risk.plot(figsize=(16, 9))
```

```python
ch_z = ch_risk.rolling(window=243).apply(lambda x: (x[-1] - x.mean()) / x.std())
```

```python
ch_z.plot(figsize=(16, 9))
```
