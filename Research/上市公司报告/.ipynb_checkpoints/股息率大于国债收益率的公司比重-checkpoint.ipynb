```python
import pyodbc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from WindPy import w
import datetime
%matplotlib inline
```

```python
w.start()
```

```python
today = datetime.datetime.today()
```

```python
start_date = '2006-01-01'
```

```python
data = w.edb('S0059752', start_date, today)
```

```python
rate = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['30y'])
```

```python
rate.index = pd.to_datetime(rate.index)
rate = rate.resample('Q').last()
```

```python
rate.tail()
```

```python
days = w.tdays(start_date, today, "Day=Alldays;Period=Q")
```

```python
dates = pd.Series(days.Data[0], index=days.Data[0])
```

```python
dates = dates.resample('Q').last()
```

```python
dates
```

### 沪深300成分股

```python
code = '000300.SH'
```

```python
percentage = pd.Series(index=dates)
for d in dates.index:
    date = dates[d]
    print(date)
    stocks = w.wset("sectorconstituent", "date=%s;windcode=%s;field=wind_code"%(date.strftime("%Y-%m-%d"), code))
    dividend_ratio = w.wss(",".join(stocks.Data[0]), "dividendyield2", "tradeDate=%s"%(date.strftime("%Y%m%d")))
    ratio = np.array(dividend_ratio.Data[0])
    ratio = ratio[ratio != 0]
    percentage[date] = (ratio > rate['30y'][d]).sum() * 1.0 / ratio.size
```

```python
data = w.wsd('000300.SH', 'close', start_date, today, "Period=M")
```

```python
close = pd.Series(data.Data[0], index=data.Times)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot(percentage, color='blue', label=u'股息率大于30年国债收益率的占比')
axx = ax.twinx()
lns2 = axx.plot(close, color='red', label=u'沪深300')

# Legend
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.07))

# Axis
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
# ax.invert_yaxis()
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals], color='blue')
axx.yaxis.set_tick_params(labelsize=15)
vals = axx.get_yticks()
axx.set_yticklabels(['{:2.0f}'.format(x) for x in vals], color='red')

_ = ax.set_title(u'沪深300成分股股息率超过30年国债收益率占比', fontsize=25, y=1.07)
# fig.savefig(u'../../picture/宏观-铜金比.jpg', format='jpg', dpi=200, bbox_inches='tight')
```
