```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.kernel_ridge import KernelRidge
from sklearn.feature_selection import SelectFromModel
import matplotlib.pyplot as plt
import datetime
import empyrical
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
df = pd.read_excel('./assets.xlsx', index_col=0)
```

```python
df.head()
```

```python
K = 121
```

```python
vol = df['065.CS'].pct_change().rolling(window=K).apply(lambda x: empyrical.annual_volatility(x))
```

```python
df['vol'] = vol
```

```python
plt.figure(figsize=(16, 9))
plt.plot(vol)
```

```python
sample = df.shift(K)
```

```python
sample['y'] = vol
```

```python
sample = sample.dropna()
```

```python
sample.head()
```

```python
split_date = '2017-09-01'
trainX, trainY = sample[sample.index < split_date][sample.columns[:-1]], sample[sample.index < split_date]['y']
testX, testY = sample[sample.index >= split_date][sample.columns[:-1]], sample[sample.index >= split_date]['y']
```

```python
lasso = linear_model.Lasso(alpha=0.0005)
lasso.fit(trainX, trainY)
```

```python
sfm = SelectFromModel(lasso, prefit=True)
```

```python
fea_trainX = sfm.transform(trainX)
fea_testX = sfm.transform(testX)
```

```python
reg_model = linear_model.LinearRegression()
# reg_model = KernelRidge(kernel='linear')
# reg_model = linear_model.Ridge()
```

```python
reg_model.fit(trainX, trainY)
```

```python
train_pred = reg_model.predict(trainX)
test_pred = reg_model.predict(testX)
```

```python
fea_futureX = df.iloc[-K:]
```

```python
future_pred = reg_model.predict(fea_futureX)
```

```python
end_date = w.tdaysoffset(K, df.index[-1]).Data[0][0]
pred_dates = w.tdays((df.index[-1] + datetime.timedelta(1)).strftime('%Y-%m-%d'), end_date.strftime('%Y-%m-%d')).Times
```

```python
plt.figure(figsize=(16, 9))
plt.plot(sample['y'], label='true')
plt.plot(pd.Series(train_pred, index=trainY.index), label='train')
plt.plot(pd.Series(test_pred, index=testY.index), label='test')
plt.plot(pd.Series(future_pred, index=pred_dates), label='future')
plt.legend()
```

```python
pd.DataFrame({'true': trainY, 'pred': train_pred}).to_csv('train.csv')
```

```python
pd.DataFrame({'true': testY, 'pred': test_pred}).to_csv('test.csv')
```

```python
pd.Series(future_pred, index=pred_dates).to_csv('future.csv')
```
