```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from WindPy import w
import datetime
%matplotlib inline
```

```python
w.start()
```

```python
start_date, end_date = '2015-02-09', datetime.datetime.today()
```

```python
data = w.edb('M0331236, M0020223', start_date, end_date)
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['vix', 'sz50'])
```

```python
df.tail()
```

```python
ret = df.pct_change()
```

```python
corr = ret['vix'].rolling(20).corr(ret['sz50'])
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 9))

ax1.plot(df['sz50'], label='上证50', color='blue')
ax1.legend(fontsize=15)
ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_tick_params(labelsize=15)

ax2.plot(df['vix'], label='波动率：50ETF期权', color='blue')
axx2 = ax2.twinx()
axx2.plot(corr, label='VIX与上证50滚动20天相关性', color='red')
ax2.legend(fontsize=15)
ax2.yaxis.set_tick_params(labelsize=15)
axx2.yaxis.set_tick_params(labelsize=15)
ax2.xaxis.set_tick_params(labelsize=15)
```
