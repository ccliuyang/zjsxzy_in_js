```python
stock = pd.read_excel('../data/基金收益.xlsx', sheet_name='股票型基金')
bond = pd.read_excel('../data/基金收益.xlsx', sheet_name='中长期纯债')
stock = stock[stock['区间收益率'] > 0]
# stock = stock[stock['年化收益率'] > 0.0]
bond = bond[bond['区间收益率'] > 0]
# bond = bond[bond['年化收益率'] > 0.0]
stock.index = np.arange(stock.shape[0])
bond.index = np.arange(bond.shape[0])
```

```python
def get_fund_returns(codes):
#     data = w.wsd(codes, 'nav_adj', start_date, end_date, 'Period=W')
#     nav = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=codes)
    jy_codes = '({})'.format(','.join(['\'{}\''.format(x.rstrip('.OF')) for x in codes]))
    query = """
            select t.SecuCode, t.SecuAbbr, nv.InfoPublDate, nv.UnitNV, nv.AccumulatedUnitNV
            from MF_NetValue as nv, SecuMain as t
            where nv.InnerCode = t.InnerCode
                and t.SecuCode in %s
                and nv.InfoPublDate >= '%s'
                and nv.InfoPublDate <= '%s'
            order by t.SecuCode, nv.InfoPublDate
            """%(jy_codes, start_date, end_date)
    nav = pd.read_sql_query(query, con=conn)
    nav = nav.set_index('InfoPublDate')
    df = pd.DataFrame()
    for code in codes:
        t = nav[nav['SecuCode'] == code.rstrip('.OF')]['AccumulatedUnitNV']
        if t.size == 0:
            return None
        if not t.index[0] == pd.to_datetime('2017-01-01'):
            return None
#         print(code, t)
        df[code] = t.resample('W').last()
    return df
```

```python
def simulation(K = 5, st_weight = 0.2, bd_weight = 0.8):
    st_index = np.random.randint(stock.shape[0], size=K)
    bd_index = np.random.randint(bond.shape[0], size=K)
    st_codes = stock.loc[st_index]['代码'].tolist()
    bd_codes = bond.loc[bd_index]['代码'].tolist()
    
    nav = get_fund_returns(st_codes + bd_codes)
    if nav is None:
        return 0, 0
    if nav.shape[1] != 2 * K:
        return 0, 0
    returns = nav.pct_change()
    weight = np.array([st_weight / K] * K + [bd_weight / K] * K)
    pf_ret = returns.mul(weight, axis=1).sum(axis=1)
    final_ret = ((1 + pf_ret).cumprod() - 1)[-1]
    an_return = (1 + final_ret) ** (1 / num_years) - 1
    an_volatility = empyrical.annual_volatility(pf_ret, period='weekly')
    
    return an_return, an_volatility
```

# 20/80组合

```python
returns2080, volatility2080 = [], []
nums = 0
for i in range(1000):
    an_return, an_volatility = simulation(K = 5)
#     print(an_return, an_volatility)
    if an_return != 0 and an_volatility != 0:
        returns2080.append(an_return)
        volatility2080.append(an_volatility)
        nums += 1
    if nums == 50:
        break
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.scatter(sd_fund['年化波动率'], sd_fund['年化收益率'], color='deepskyblue', label='二级债基')
ax.scatter(mix_fund['年化波动率'], mix_fund['年化收益率'], color='blue', label='偏债混合型基金')
# ax.scatter(lb_fund['年化波动率'], lb_fund['年化收益率'], color='cyan', label='中长期纯债基金')
# ax.scatter(st_fund['年化波动率'], st_fund['年化收益率'], color='blueviolet', label='股票/偏股型基金')
ax.scatter(volatility2080, returns2080, color='red', label='20/80基金组合')

ax.legend(fontsize=15)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.1f}%'.format(x*100) for x in vals])
vals = ax.get_xticks()
ax.set_xticklabels(['{:2.1f}%'.format(x*100) for x in vals])
ax.set_xlabel('年化波动率', fontsize=25)
ax.set_ylabel('年化收益率', fontsize=25)
```
