```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import empyrical
%matplotlib inline
```

```python
df = pd.read_excel('../data/us & china stock band data.xlsx')
```

```python
df.tail()
```

```python
ret = df.pct_change()
```

```python
(1 + ret).cumprod().plot(figsize=(16, 9))
```

```python
for col in ret.columns:
    print(col, empyrical.annual_return(ret[col].dropna()), empyrical.annual_volatility(ret[col].dropna()))
```

```python
ret['美股'].corr(ret['美债'])
```

```python
ret['A股'].corr(ret['中债'])
```

```python
def _allocation_risk(weights, covariances):
    
    # calculate the risk of the weights distribution
    portfolio_risk = np.sqrt((weights * covariances * weights.T))[0, 0]
    
    # It returns the risk of the weights distribution
    return portfolio_risk
```

# 美股60/40组合波动率

```python
weights = np.array([0.6, 0.4])
covariance = ret[['美股', '美债']].cov()
```

```python
us_risk = _allocation_risk(weights, covariance.values)
```

```python
us_risk
```

```python
china_risk = []
covariance = ret[['A股', '中债']].cov()
for w in range(100):
    w *= 0.01
    risk = _allocation_risk(np.array([w, 1-w]), covariance.values)
    china_risk.append(risk)
```

```python
diff = []
for i, v in enumerate(china_risk):
    diff.append(abs(v - us_risk))
diff = np.array(diff)
```

```python
diff.argmin()
```

### 结论：38%股票，62%债券

# 美股60/40组合收益率

```python
us_stock_ret = empyrical.annual_return(ret['美股'].dropna())
us_bond_ret = empyrical.annual_return(ret['美债'].dropna())
us_pf_ret = 0.6 * us_stock_ret + 0.4 * us_bond_ret
```

```python
us_pf_ret
```

```python
cn_stock_ret = empyrical.annual_return(ret['A股'].dropna())
cn_bond_ret = empyrical.annual_return(ret[ret.index >= '2005-01-01']['中债'].dropna())
```

```python
cn_stock_ret, cn_bond_ret
```

```python
china_ret = []
for w in range(100):
    w *= 0.01
    r = w * cn_stock_ret + (1 - w) * cn_bond_ret
    china_ret.append(r)
```

```python
diff = []
for i, v in enumerate(china_ret):
    diff.append(abs(v - us_pf_ret))
diff = np.array(diff)
```

```python
diff.argmin()
```

### 结论：44%股票，56%债券
