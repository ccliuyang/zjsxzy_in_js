```python
import pandas as pd
import numpy as np
from WindPy import w
import empyrical
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
def metric(daily_ret):
    tot_ret = empyrical.cum_returns_final(daily_ret)
    an_ret = empyrical.annual_return(daily_ret)
    sharpe = empyrical.sharpe_ratio(daily_ret, risk_free=0.015/365)
    vol = empyrical.annual_volatility(daily_ret)
    maxdraw = empyrical.max_drawdown(daily_ret)
    print('Total return: %.2f%%'%(tot_ret*100))
    print('Annual return: %.2f%%'%(an_ret*100))
    print('Sharpe ratio: %.2f'%(sharpe))
    print('Annual volatility: %.2f%%'%(vol*100))
    print('Max drawdown: %.2f%%'%(maxdraw*100))
```

```python
df = pd.read_excel(u'../data/20200114客户抽样.xlsx', 
                   sheet_name=u'客户6', 
                   skiprows=17,
                   converters = {u'基金代码': str})
```

```python
df[u'确认日期'] = pd.to_datetime(df[u'确认日期'])
df[u'申请日期'] = pd.to_datetime(df[u'申请日期'])
```

# 创建组合

```python
portfolio = pd.DataFrame(index=nav_adj.index, columns=nav_adj.columns) # 持有组合基金金额
portfolio_shares = pd.DataFrame(index=nav_adj.index, columns=nav_adj.columns) # 持有组合基金份额
current_porfolio, current_shares = {}, {} # 当前持有基金金额、当前持有基金份额

for index in df.index:
    date = df.loc[index][u'申请日期']
    fund = '%s.OF'%(df.loc[index][u'基金代码'])
    share = df.loc[index][u'确认份额']
    invest = df.loc[index][u'确认金额']
    ftype = df.loc[index][u'业务类型']
    
    if ftype == u'申购' or ftype == u'基金成立' or ftype == u'定期定额申购' or ftype == u'基金转换入':
        if fund not in current_porfolio:
            current_porfolio[fund] = 0
            current_shares[fund] = 0
        current_porfolio[fund] += invest
        current_shares[fund] += share
    elif ftype == u'赎回' or ftype == u'快速取现' or ftype == u'基金转换(出)':
        if fund in current_porfolio:
            invest = min(invest, current_porfolio[fund])
            share = min(share, current_shares[fund])
            current_porfolio[fund] -= invest
            current_shares[fund] -= share
    
    for k, v in current_porfolio.items():
        portfolio.loc[date][k] = v
    for k, v in current_shares.items():
        portfolio_shares.loc[date][k] = v
    
#     print(ftype)
#     print(date, current_porfolio)
#     print(portfolio.loc[date])
```

```python
portfolio_shares = portfolio_shares.fillna(method='ffill').fillna(0)
portfolio = portfolio.fillna(method='ffill').fillna(0) # 成本
```

```python
returns = (portfolio.div(portfolio.sum(axis=1), axis=0) * nav_adj.pct_change()).sum(axis=1) # 按初始金额加权的投资收益（估算）
```
