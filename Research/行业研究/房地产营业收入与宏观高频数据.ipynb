```python
import pandas as pd
import numpy as np
from WindPy import w
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
df = pd.read_excel(u'./data/房地产.xlsx')
```

```python
df.tail()
```

```python
data = w.edb('S0073300,M0001383', '2004-03-01', '2019-07-05')
```

```python
econ = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['m1', 'sale'])
```

```python
econ.tail()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot(df[u'营业收入YOY'], color='blue', label=u'营业收入YOY')
axx = ax.twinx()
lns2 = axx.plot(econ[u'sale'].dropna(), color='red', label=u'商品房销售面积：累计同比')
# lns3 = axx.plot(econ['m1'].dropna(), color='black', label=u'M1：同比')

# zero line
axx.axhline(y=0, color='black', linewidth=1)

# Legend
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.07))

# Axis
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals], color='blue')
axx.yaxis.set_tick_params(labelsize=15)
vals = axx.get_yticks()
axx.set_yticklabels(['{:2.0f}%'.format(x) for x in vals], color='red')

_ = ax.set_title(u'房地产营业收入与商品房销售面积', fontsize=25, y=1.07)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot(df[u'营业收入YOY'], color='blue', label=u'营业收入YOY')
axx = ax.twinx()
# lns2 = axx.plot(econ[u'sale'].dropna(), color='red', label=u'商品房销售面积：累计同比')
lns2 = axx.plot(econ['m1'].dropna(), color='red', label=u'M1：同比')

# zero line
axx.axhline(y=0, color='black', linewidth=1)

# Legend
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.07))

# Axis
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals], color='blue')
axx.yaxis.set_tick_params(labelsize=15)
vals = axx.get_yticks()
axx.set_yticklabels(['{:2.0f}%'.format(x) for x in vals], color='red')

_ = ax.set_title(u'房地产营业收入与M1', fontsize=25, y=1.07)
```
