```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from WindPy import w
import datetime
%matplotlib inline
```

```python
w.start()
```

```python
today = datetime.datetime.today()
```

```python
start_date = '2010-01-01'
```

```python
data = w.edb('M0020209,M0062541,S0059749', start_date, today)
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['300', '500', '10y'])
```

```python
df['ret'] = df['300'].pct_change() - df['500'].pct_change()
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot((1 + df['ret']).cumprod().dropna(), color='blue')
axx = ax.twinx()
lns2 = axx.plot(df['10y'], color='red')

# Legend
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.07))

# Axis
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
axx.yaxis.set_tick_params(labelsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.2f}'.format(x) for x in vals], color='blue')
vals = axx.get_yticks()
axx.set_yticklabels(['{:2.2f}%'.format(x) for x in vals], color='red')
```
