# 标准风险平价

```python
 # 风险预算优化
def calculate_portfolio_var(w, V):
    # 计算组合风险的函数
    w = np.matrix(w)
    return (w * V * w.T)[0, 0]
```

```python
def calculate_risk_contribution(w ,V):
    # 计算单个资产对总体风险贡献度的函数
    w = np.matrix(w)
    sigma = np.sqrt(calculate_portfolio_var(w, V))
    # 边际风险贡献
    MRC = V * w.T
    # 风险贡献
    RC = np.multiply(MRC, w.T) / sigma
    return RC
```

```python
def risk_budget_objective(x, pars):
    # 计算组合风险
    V = pars[0]# 协方差矩阵
    x_t = pars[1] # 组合中资产预期风险贡献度的目标向量
    sig_p =  np.sqrt(calculate_portfolio_var(x, V)) # portfolio sigma
    risk_target = np.asmatrix(np.multiply(sig_p, x_t))
    asset_RC = calculate_risk_contribution(x, V)
    J = sum(np.square(asset_RC - risk_target.T))[0, 0] # sum of squared error
#     J = np.abs(asset_RC.max() - asset_RC.min())
    return J
```

```python
def total_weight_constraint(x):
    return np.sum(x) - 1.0
```

```python
def long_only_constraint(x):
    return x
```

```python
# 根据资产预期目标风险贡献度来计算各资产的权重
def calcu_w(x, V):
    w0 = [0.1, 0.1, 0.1, 0.3, 0.2, 0.2]
#     w0 = [0.1, 0.1, 0.1, 0.7] 
#     x_t = [0.25, 0.25, 0.25, 0.25] # 目标是让四个资产风险贡献度相等，即都为25%
    x_t = x
    cons = ({'type': 'eq', 'fun': total_weight_constraint},
    {'type': 'ineq', 'fun': long_only_constraint})
    res = minimize(risk_budget_objective, w0, args=[V, x_t], method='SLSQP',constraints=cons)
    print res
    w_rb = np.asmatrix(res.x)
    return w_rb
```

```python
def metric(daily_ret):
    an_ret = pf.empyrical.annual_return(daily_ret)
    sharpe = pf.empyrical.sharpe_ratio(daily_ret)
    vol = pf.empyrical.annual_volatility(daily_ret)
    maxdraw = pf.empyrical.max_drawdown(daily_ret)
    print 'Annual return: %.2f%%'%(an_ret*100)
    print 'Sharpe ratio: %.2f'%(sharpe)
    print 'Annual volatility: %.2f%%'%(vol*100)
    print 'Max drawdown: %.2f%%'%(maxdraw*100)
```

## 另一种cost function
$$J = \|max(rc) - min(rc) \|$$

```python
def risk_budget_objective2(x, pars):
    # 计算组合风险
    V = pars[0]# 协方差矩阵
    x_t = pars[1] # 组合中资产预期风险贡献度的目标向量
    sig_p =  np.sqrt(calculate_portfolio_var(x, V)) # portfolio sigma
    risk_target = np.asmatrix(np.multiply(sig_p, x_t))
    asset_RC = calculate_risk_contribution(x, V)
    J = np.abs(asset_RC.max() - asset_RC.min())
    return J
```

```python
# 根据资产预期目标风险贡献度来计算各资产的权重
def calcu_w2(x, V):
    w0 = [0.1, 0.1, 0.1, 0.3, 0.2, 0.2]
#     w0 = [0.1, 0.1, 0.1, 0.7] 
#     x_t = [0.25, 0.25, 0.25, 0.25] # 目标是让四个资产风险贡献度相等，即都为25%
    x_t = x
    cons = ({'type': 'eq', 'fun': total_weight_constraint},
    {'type': 'ineq', 'fun': long_only_constraint})
    res = minimize(risk_budget_objective2, w0, args=[V, x_t], method='SLSQP',constraints=cons)
    print res
    w_rb = np.asmatrix(res.x)
    return w_rb
```

## Empirical covariance

```python
weight = pd.DataFrame(columns=new_df.columns, index=new_df.index)
wb = np.array([0, 0, 0, 0, 0, 0])
ec = EmpiricalCovariance(assume_centered=True)
for date in new_df.index:
    weight.loc[date] = wb
    if date in rebalance_dates:
        X_train = ret_df[ret_df.index <= date][-60:]
        ec.fit(X_train)
        wb = calcu_w([1./6]*6, ec.covariance_*10000)
        wb = np.array(wb)[0]
```

```python
ec_portfolio_ret = (weight * ret_df).sum(axis=1)
```

```python
metric(ec_portfolio_ret)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+equal_portfolio_ret).cumprod(), label='equal weight')
plt.plot((1+sample_portfolio_ret).cumprod(), label='sample covariance risk parity')
plt.plot((1+ec_portfolio_ret).cumprod(), label='ec covariance risk parity')
plt.legend()
```
