```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pnl = pd.read_pickle('D:/Data/price.pkl')
```

```python
df = pnl.minor_xs('close').pct_change().abs() / pnl.minor_xs('volume')
```

```python
liquidity = df.median(axis=1)
```

```python
plt.figure(figsize=(16, 9))
liquidity.plot()
```
