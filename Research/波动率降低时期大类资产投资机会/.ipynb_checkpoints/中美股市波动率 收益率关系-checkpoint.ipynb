```python
import pandas as pd
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import empyrical
%matplotlib inline
```

```python
sns.set(style="white", context="talk")
```

```python
def roll(df, w):
    """
    This fucntion comes from:
    http://stackoverflow.com/questions/37486502/why-does-pandas-rolling-use-single-dimension-ndarray/37491779#37491779
    """
    df.fillna(df.mean(), inplace=True)
    roll_array = np.dstack([df.values[i:i+w, :] for i in range(len(df.index) - w + 1)]).T
    panel = pd.Panel(roll_array,
                     items=df.index[w-1:],
                     major_axis=df.columns,
                     minor_axis=pd.Index(range(w), name='roll'))
    return panel.to_frame().unstack().T.groupby(level=0)
```

```python
df = pd.read_excel('./data/cn_stock.xlsx', index_col=0)
```

```python
df.head()
```

```python
K = 242
```

```python
df['return'] = df['000001.SH'].pct_change().rolling(window=K).apply(lambda x: empyrical.annual_return(x))
```

```python
df['volatility'] = df['000001.SH'].pct_change().rolling(window=K).apply(lambda x: empyrical.annual_volatility(x))
```

```python
df['sharpe'] = df['000001.SH'].pct_change().rolling(window=K).apply(lambda x: empyrical.sharpe_ratio(x))
```

```python
sample = df.resample('M').last()
```

```python
# sample.to_csv('./data/cn_output.csv')
```

```python
sample = sample[sample.index >= '1992-05-31']
```

# 股市波动率

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
plt.bar(sample.index, sample['volatility'])
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
```

```python
roll_df = roll(df, 242)
```

```python
df = df[df.index >= '2000-01-01']
```

```python
quantiles = df.quantile(np.arange(0., 1.1, 0.1))['volatility']
```

```python
ret = pd.Series(index=range(quantiles.shape[0] - 1))
for i, q in enumerate(quantiles):
    if i != 0:
        avg_ret = df[(df['volatility'] >= quantiles.iloc[i-1]) & (df['volatility'] <= quantiles.iloc[i])]['sharpe'].mean()
        print avg_ret
        ret.iloc[i-1] = avg_ret
ret = pd.DataFrame(ret, columns=['sharpe'])
ret['group'] = ['Group %d'%(i+1) for i in range(ret.shape[0])]
```

```python
fig = plt.figure(figsize=(16, 9))
sns.set_style("whitegrid")
ax = fig.add_subplot(111)
sns.barplot('group', 'sharpe', data=ret, palette="Blues_d")
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
```
