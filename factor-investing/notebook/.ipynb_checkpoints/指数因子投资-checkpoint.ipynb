# 选出的资产做风险平价配置

```python
weight_df = pd.DataFrame(columns=pnl.minor_axis, index=pnl['momentum per'].dropna().index)
```

```python
 # 风险预算优化
def calculate_portfolio_var(w, V):
    # 计算组合风险的函数
    w = np.matrix(w)
    return (w * V * w.T)[0, 0]
```

```python
def calculate_risk_contribution(w ,V):
    # 计算单个资产对总体风险贡献度的函数
    w = np.matrix(w)
    sigma = np.sqrt(calculate_portfolio_var(w, V))
    # 边际风险贡献
    MRC = V * w.T
    # 风险贡献
    RC = np.multiply(MRC, w.T) / sigma
    return RC
```

```python
def risk_budget_objective(x, pars):
    # 计算组合风险
    V = pars[0]# 协方差矩阵
    x_t = pars[1] # 组合中资产预期风险贡献度的目标向量
    sig_p =  np.sqrt(calculate_portfolio_var(x, V)) # portfolio sigma
    risk_target = np.asmatrix(np.multiply(sig_p, x_t))
    asset_RC = calculate_risk_contribution(x, V)
    J = sum(np.square(asset_RC - risk_target.T))[0, 0] # sum of squared error
#     J = np.abs(asset_RC.max() - asset_RC.min())
    return J
```

```python
def total_weight_constraint(x):
    return np.sum(x) - 1.0
```

```python
def long_only_constraint(x):
    return x
```

```python
# 根据资产预期目标风险贡献度来计算各资产的权重
def calcu_w(x, V):
    w0 = np.ones(len(x))
#     w0 = [0.1, 0.1, 0.1, 0.7] 
#     x_t = [0.25, 0.25, 0.25, 0.25] # 目标是让四个资产风险贡献度相等，即都为25%
    x_t = x
    cons = ({'type': 'eq', 'fun': total_weight_constraint},
    {'type': 'ineq', 'fun': long_only_constraint})
    res = minimize(risk_budget_objective, w0, args=[V, x_t], method='SLSQP',constraints=cons)
    print res
    w_rb = np.asmatrix(res.x)
    return w_rb
```

```python
weight = pd.Series(0, index=pnl.minor_axis)
k = 3
for date in weight_df.index:
    weight_df.loc[date] = weight
    if date in rebanlance_dates:
        # rebalance
        weight = pd.Series(0, index=pnl.minor_axis)
        df = pnl.major_xs(date)
        
        # carry
        weight[df['carry per'].sort_values()[:k].index] += 1
        # value
        weight[df['value per'].sort_values()[:k].index] += 1
        # defensive
        weight[df['defensive'].sort_values()[:k].index] += 1
        # momentum
        weight[df['momentum'].sort_values(ascending=False)[:5].index] += 1
        
        weight /= weight.sum()
#         print date
#         print weight[weight != 0]
        assets = weight[weight != 0].index
#         print assets
        n = assets.shape[0]
        df = pnl['return'][assets]
        cov = df[df.index <= date][-25:].cov()
        wb = calcu_w([1./n]*n, (cov*10000).as_matrix())
        wb = np.array(wb)[0]
        weight[assets] = wb
#         break
#         print weight
```

```python
ret = (weight_df * pnl['return'].loc[weight_df.index]).sum(axis=1)
```

```python
metric(ret)
```
